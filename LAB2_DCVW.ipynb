```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

In this assignment you need to work with data from the [worldometers](https://www.worldometers.info/coronavirus/) website. I want you to scrape all 215 countries information about coronavirus cases from the website.** 
The data has to include:
- `Country name`
- `Total cases`
- `Total deaths`
- `Total recovered`
- `Active cases`
- `New cases`
- `New deaths`
- `Total tests`
- `Population`

```python
url = "https://www.worldometers.info/coronavirus/"
data = requests.get(url).text
soup = BeautifulSoup(data, 'html.parser')
```

```python
import re
table = soup.find('table', id='main_table_countries_today')
df = pd.DataFrame(columns=['Country name', 'Total cases', 'Total deaths', 'Total recovered', 'Active cases', 'New cases', 'New deathes', 'Total tests', 'Population'])

# table.tbody.find_all('tr', class_ = re.compile(r"^(even|odd)$"))
# I actually want to use a regular expression but can't implement it, try using the above code to search for even/odd tr-s 

for row in table.tbody.find_all('tr', class_= lambda c: not c or 'even' in c or 'odd' in c):  
    columns = row.find_all('td')
    if(columns != []):
        country_name = columns[1].text.strip()
        total_cases = columns[2].text.strip()
        total_deaths= columns[4].text.strip()
        total_recovered = columns[6].text.strip()
        active_cases = columns[8].text.strip()
        new_cases = columns[3].text.strip()
        new_deathes = columns[5].text.strip()
        total_tests = columns[12].text.strip()
        population = columns[14].text.strip()
        
        df = df.append({'Country name': country_name,  'Total cases': total_cases, 'Total deaths': total_deaths, 'Total recovered': total_recovered, 'Active cases': active_cases, 'New cases': new_cases, 'Total tests': total_tests, 'Population': population}, ignore_index=True)

df.to_csv('data.csv')
```
