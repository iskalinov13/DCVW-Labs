```python
# 1. Import needed libraries and download data from here:
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount("/drive") 
```

```python
# 2. Save the data as df with name users
PATH = '/drive/My Drive/Colab Notebooks/data.csv'
data = pd.read_csv(PATH)
data.head()
```

```python
# 3. Discover what is the mean age per occupation 
data.groupby('occupation')['age'].mean()
```

```python
# 4. Discover the Male ratio per occupation and sort it from the most to the least
data['gender_m'] = data['gender'].apply(lambda x: 1 if x == 'M' else 0)
res = (data.groupby('occupation')['gender_m'].sum() / data['occupation'].value_counts() * 100).sort_values(ascending = False)
res
```

```python
# 5. For each occupation, calculate the minimum and maximum ages
data.groupby('occupation')['age'].describe()[['min','max']]
```

```python
# 6. For each combination of occupation and gender, calculate the mean age
data.groupby(['occupation', 'age'])['age'].mean()
```

```python
# 7. For each occupation present the percentage of women and men
gender_ocup = data.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count = data.groupby(['occupation']).agg('count')
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100
occup_gender.loc[: , 'gender']
```
