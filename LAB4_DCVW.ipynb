```python
# for this assignment we will use ready datasets from sklearn laibrary
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
data = load_iris()
# 1. How this data is stored? 
```

```python
col_names = data['feature_names']
col_names
targets = data['target']
data = data['data']
col_names
```

```python
# 2.report sizes of col_names, data and targets
print("Size of col_names is:", len(col_names))
print("Size of data is:", data.shape)
print("Size of targets is:", targets.shape)
```

```python
# 3. Create a dataframe from data and name it as df
df = pd.DataFrame(data)
df
```

```python
# 4. Name the column names as col_names in df. Example, the first column name should be 'sepal length'
df.columns = col_names
df
```

```python
# 5. Add another column to df named 'Target' with the values corresponding to targets
df['Target'] = targets
df
```

```python
# 6. Report the size of the dataframe
df.shape
```
